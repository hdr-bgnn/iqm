Generate metadata

```python
meta=pd.read_csv("../../data/INHS_corrected_2020.06.09.csv")

#annotated webform data
qual=pd.read_csv("../../data/hdrwebform_20200728.csv") 
meta
#df['Discounted_Price'] = df.apply(lambda row: row.Cost - (row.Cost * 0.1), axis = 1) 

meta['fileexists']=meta.apply(lambda x: \
                              os.path.exists(os.path.join("../../data/INHS_segmented_padded_fish/images",x['fileName'])),axis=1)
qual=qual.drop(columns=["fixed_scientific_name","scientific_name","resolution"]) #all NAN
moige=qual.set_index('image_name').join(meta.set_index('fileName'),how='inner')
#drop dups and nulls
moige=moige.reset_index().rename(columns={'index':'fileName'}).drop_duplicates()
moige=moige[moige.scientificName.notna()]
#fileName	scientificName	Genus	Family
#INHS_FISH_59772.jpg	Lepomis cyanellus	Lepomis	Centrarchidae
#INHS_FISH_4051.jpg	Noturus gyrinus	Noturus	Ictaluridae
#INHS_FISH_38963.jpg	Esox americanus	Esox	Esocidae
```

```python
#moige.groupby('scientificName').agg('size')[moige.groupby('scientificName').agg('size')>10].keys()
#just the scinames withmore than 10 instances
scinames=moige.groupby('scientificName').agg('size')[moige.groupby('scientificName').agg('size')>20].keys()
moige=moige[moige.scientificName.isin(scinames)]
```

```python
moige[moige.fileexists==False]['fileName']
```

```python
moige=moige[moige.fileexists==True]
#big_genera=moige.groupby('Genus').agg('size')[moige.groupby('Genus').agg('size')>10].keys()
moige[['fileName','scientificName','Genus','Family']].to_csv("../../data/allgraded.csv",sep="\t",index=False)
```
